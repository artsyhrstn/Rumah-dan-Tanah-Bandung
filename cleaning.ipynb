```python
import json
import requests
import pandas as pd
```

```python
with open(r'.\\data\\tanah.json', 'r', encoding='utf-8') as file:
    tanah_data = json.load(file)
print(tanah_data)
```

```python
from pandas import json_normalize

# Normalize the 'data' field in tanah_data
normalized_data = json_normalize(tanah_data['data'])
print(normalized_data.head())
```

```python
normalized_data.to_csv(r'.\\data\\tanah.csv', index=False)
```
